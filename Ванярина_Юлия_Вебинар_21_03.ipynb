```python
import pandas as pd
# Загружаем данные
df = pd.read_csv('joined_data.csv')
```

```python
df
```

```python
import pandas as pd
import plotly.express as px

first_day_temp = df.iloc[0]['temperature']
last_day_temp = df.iloc[-1]['temperature']

print(f"Температура первого дня: {first_day_temp}°C")
print(f"Температура последнего дня: {last_day_temp}°C")

if first_day_temp > last_day_temp:
    print("Первый день теплее.")
elif first_day_temp < last_day_temp:
    print("Последний день теплее.")
else:
    print("Температуры первого и последнего дня равны.")

fig = px.line(df, x='date', y='temperature', title='Температура по дням',
              labels={'date': 'Дата', 'temperature': 'Температура (°C)'},
              text='temperature', markers=True)

# Настраиваем отображение значений при наведении
fig.update_traces(textposition='top center', hovertemplate='Дата: %{x}<br>Температура: %{y}°C')

# Настраиваем оси и общий вид графика
fig.update_layout(
    xaxis_title='Дата',
    yaxis_title='Температура (°C)',
    hovermode='x unified',  # Показывать значения при наведении на все линии
    showlegend=False,       # Скрыть легенду (если не нужна)
    template='plotly_white' # Стиль графика
)

# Показываем график
fig.show()
```
